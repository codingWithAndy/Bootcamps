## Example 1 - Simple Cost Function

$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start = -3, stop = 3, num = 500)
x_1
```

```python
# Plot
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel("X", fontsize = 16)
plt.ylabel("f(x)", fontsize = 16)

plt.plot(x_1, f(x_1))
plt.show()
```

## Slope and Derivatives

Challenge: Create a Python file for the derivative of $f(x)$ called $df(x)$ 

```python
def df(x):
    return 2 * x + 1
```

```python
# Plot function and Derivative side by side

plt.figure(figsize = [15, 5])

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim([-3, 3])
plt.ylim(0, 8)

plt.title("Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("f(x)", fontsize = 16)

plt.plot(x_1, f(x_1), color = "blue", linewidth = 3)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title("Slope of Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("df(x)", fontsize = 16)

plt.grid()

plt.xlim([-2, 3])
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color = "skyblue", linewidth = 5)

plt.show()
```

```python
# Plot function and Derivative above each other.

plt.figure(figsize = [15, 5])

# 1 Chart: Cost Function
plt.subplot(2, 1, 1)

plt.xlim([-3, 3])
plt.ylim(0, 8)

plt.title("Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("f(x)", fontsize = 16)

plt.plot(x_1, f(x_1), color = "blue", linewidth = 3)

# 2 Chart: Derivative
plt.subplot(2, 1, 2)

plt.title("Slope of Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("df(x)", fontsize = 16)

plt.grid()

plt.xlim([-2, 3])
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color = "skyblue", linewidth = 5)

plt.show()
```

## Python Loops & Gradient Descent

```python
# Python For Loop
for n in range(5):
    print("Hello World!", n)

print("End of Loop")
```

```python
# Python While Loop
counter = 0

while counter < 7:
    print("counting.....", counter)
    counter += 1

print("Ready or not, here I come!")
```

```python
# Gradient Descent

new_x = -3 #New Starting point
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

# Variables for graphing diagram
x_list = [new_x]
slope_list = [df(previous_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print("Loop ran this many times:", n)
        break
    
print("Local minimum occurs at:", new_x)
print("Slope or df(x) at this point is:", df(new_x))
print("f(x) calue or cost at this point is:", f(new_x))
    
```

```python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
```

# Example 2 - Multiple Minima vs Initial Guess & Advanced Functions

## $$g(x) = x^4 - 4x^2 + 5$$

```python
# Make some data
x_2 = np.linspace(-2, 2, 1000)

# Challenge: Write the g(x) function and the dg(x) function in Python.
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4 * x**3 - 8*x
```

```python
# Plot function and Derivative side by side

plt.figure(figsize = [15, 5])

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)

plt.title("Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("g(x)", fontsize = 16)

plt.plot(x_2, g(x_2), color = "blue", linewidth = 3)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title("Slope of Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("dg(x)", fontsize = 16)

plt.grid()

plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.plot(x_2, dg(x_2), color = "skyblue", linewidth = 5)

plt.show()
```

## Gradient Descent as a Python Function

```python
def gradient_descent(deriviative_func, initial_guess, multiplier, precision):

    new_x = initial_guess
    x_list = [new_x]
    slope_list = [deriviative_func(new_x)]

    for n in range(500):
        previous_x = new_x
        gradient = deriviative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        slope_list.append(deriviative_func(new_x))

        if step_size < precision:
            break

    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)

print("Local min occurs at:", local_min)
print("Number of steps:", len(list_x))
```

```python
# Alternative to the above code -> uses peramitar names for reference.
local_min, list_x, deriv_list = gradient_descent(deriviative_func= dg, initial_guess= -0.5, multiplier= 0.02, 
                                                 precision= 0.001)

print("Local min occurs at:", local_min)
print("Number of steps:", len(list_x))
```

```python
# Option peramitar version (Multiplier and Precision are now optional).
def gradient_descent2(deriviative_func, initial_guess, multiplier= 0.02, precision= 0.001, max_inter= 300):

    new_x = initial_guess
    x_list = [new_x]
    slope_list = [deriviative_func(new_x)]

    for n in range(max_inter):
        previous_x = new_x
        gradient = deriviative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        slope_list.append(deriviative_func(new_x))

        if step_size < precision:
            break

    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent2(deriviative_func= dg, initial_guess= -0.1)

print("Local min occurs at:", local_min)
print("Number of steps:", len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent2(deriviative_func= dg, initial_guess= -0.5, multiplier= 0.01, 
                                                 precision= 0.0001)
```

```python
# Calling Gradient descent function

local_min, list_x, deriv_list = gradient_descent2(deriviative_func= dg, initial_guess=0)

# Plot function and Derivative and scatter plot side by side

plt.figure(figsize = [15, 5])

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)

plt.title("Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("g(x)", fontsize = 16)

plt.plot(x_2, g(x_2), color = "blue", linewidth = 3, alpha= 0.8)
plt.scatter(list_x, g(np.array(list_x)), color= "red", s= 100, alpha= 0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title("Slope of Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("dg(x)", fontsize = 16)

plt.grid()

plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.plot(x_2, dg(x_2), color = "skyblue", linewidth = 5, alpha= 0.6)
plt.scatter(list_x, deriv_list, color= "red", s=100, alpha= 0.5)

plt.show()
```

# Example 3 - Divergence, Overflow and Python Tuples

## $$h(x) = x^5 - 2x^4 + 2$$

```python
# Make data
x_3 = np.linspace(start= -2.5, stop= 2.5, num= 1000)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5 * x**4 - 8*x**3
```

```python
# Calling Gradient descent function

local_min, list_x, deriv_list = gradient_descent2(deriviative_func= dh, initial_guess= -0.2, max_inter= 70)

# Plot function and Derivative and scatter plot side by side

plt.figure(figsize = [15, 5])

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)

plt.title("Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("h(x)", fontsize = 16)

plt.plot(x_3, h(x_3), color = "blue", linewidth = 3, alpha= 0.8)
plt.scatter(list_x, h(np.array(list_x)), color= "red", s= 100, alpha= 0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title("Slope of Cost Function", fontsize = 17)
plt.xlabel("X", fontsize = 16)
plt.ylabel("dh(x)", fontsize = 16)

plt.grid()

plt.xlim(-1, 2)
plt.ylim(-4, 5)

plt.plot(x_3, dh(x_3), color = "skyblue", linewidth = 5, alpha= 0.6)
plt.scatter(list_x, deriv_list, color= "red", s=100, alpha= 0.5)

plt.show()

print("Local min occurs at:", local_min)
print("Cost of this minimum is:", h(local_min))
print("Number of steps:", len(deriv_list))
```
